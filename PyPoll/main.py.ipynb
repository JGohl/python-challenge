```python
import os
import csv

poll_data = os.path.join("..\PyPollResources", "election_data.csv")

with open (poll_data, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader, None)
    vote_count = 0
    khan_count = 0
    correy_count = 0
    li_count = 0
    otooley_count = 0
    other_count = 0
    winner = str
    
    for row in csvreader:
        vote_count += 1
        if row[2] == "Khan":
            khan_count += 1
        elif row[2] == "Correy":
            correy_count += 1
        elif row[2] == "Li":
            li_count += 1
        elif row[2] == "O'Tooley":
            otooley_count += 1
        else:
            other_count += 1
    khan_perc = round((float(khan_count / vote_count) * 100), 3)
    correy_perc = round((float(correy_count / vote_count) * 100), 3)
    li_perc = round((float(li_count / vote_count) * 100), 3)
    otooley_perc = round((float(otooley_count / vote_count) * 100), 3)
    
    if khan_count > (correy_count and li_count and otooley_count):
        winner = "Khan"
    elif correy_count > (khan_count and li_count and otooley_count):
        winner = "Correy"
    elif li_count > (khan_count and correy_count and otooley_count):
        winner = "Li"
    else:
        winner = "O'Tooley"
    
        

print("Election Results")
print("---------------------------")
print("Total Votes: " + str(vote_count))
print("---------------------------")
print("Khan: " + format(khan_perc, ".3f") + "% (" +str(khan_count) + ")")
print("Correy: " + format(correy_perc, ".3f") + "% (" +str(correy_count) + ")")
print("Li: " + format(li_perc, ".3f") + "% (" +str(li_count) + ")")
print("O'Tooley: " + format(otooley_perc, ".3f") + "% (" +str(otooley_count) + ")")
print("---------------------------")
print("Winner: " + winner)
print("---------------------------")


f = open("PyPollOutput.txt", "w")

print("Election Results", file=f)
print("---------------------------", file=f)
print("Total Votes: " + str(vote_count), file=f)
print("---------------------------", file=f)
print("Khan: " + format(khan_perc, ".3f") + "% (" +str(khan_count) + ")", file=f)
print("Correy: " + format(correy_perc, ".3f") + "% (" +str(correy_count) + ")", file=f)
print("Li: " + format(li_perc, ".3f") + "% (" +str(li_count) + ")", file=f)
print("O'Tooley: " + format(otooley_perc, ".3f") + "% (" +str(otooley_count) + ")", file=f)
print("---------------------------", file=f)
print("Winner: " + winner, file=f)
print("---------------------------", file=f)

f.close()
```
